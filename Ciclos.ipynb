```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('base.csv', index_col='N')
df.head() #Tail para ver las ultimas filas
```

```python
#Estadisticos
df.describe()
```

```python
#Limpieza de datos
df_filtrado = df.fillna(0)   # df.fillna({'inflacion': 0, 'Tasas': -1})
df_filtrado.head()
```

```python
#Plot onlyone
plt.figure(figsize=(9, 4))
plt.plot(df_filtrado['EFFR'])


plt.xlabel('Tiempo')
plt.ylabel('Porsentaje')
plt.title('Tasa de interes')


plt.show()
```

```python
#Plot two

plt.figure(figsize=(9, 4))
plt.plot(df_filtrado['CPIAUCSL Indexada'])
plt.plot(df_filtrado['EFFR Indexada'])


plt.xlabel('Tiempo')
plt.ylabel('Porsentaje')
plt.title('Tasa de interes')


plt.show()
```
